Now lets see how these each statistic correlates with each other

```python
plt.figure(figsize=(12, 10))
corr = df[['Distance (ly)', 'Luminosity (L/Lo)', 'Radius (R/Ro)', 'Temperature (K)']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

From this we can see that there is a strong positive correlation between distance and luminosity, this tells us that the greater the distance a star is away from earth, the greater the luminosity of said star is likely to be.
This is likely because we can only see stars with a lower luminosity which are closer and that we lose visibility of them the further away they are.

We can also see that there is a marginal positive correlation between radius and luminosity, which impies that bigger stars may be more luminous than smaller stars.

Now we will plot all star statistics against one another, this will allow us to identify any potential outliers and inform us on areas to investigate further.

```python
sns.pairplot(df[['Distance (ly)', 'Luminosity (L/Lo)', 'Radius (R/Ro)', 'Temperature (K)']])
plt.suptitle('Pair Plot of Star Parameters', y=1.02)
plt.show()
```
